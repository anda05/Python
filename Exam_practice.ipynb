### 32 a - button show text

```python
from tkinter import *
w=Tk()
w.geometry('500x500')
w.title("Click a button to show a message")
def show_bob():
    label.config(text='Hello Bob')
def show_alice():
    label.config(text='Hello Alice')
label=Label(w,text='',font=('Arial',20),fg='magenta')
label.pack(pady=10)
b1=Button(w,text='Bob',font='Arial', width=10, height=4,activebackground='lightblue',activeforeground='blue',command=show_bob)
b2=Button(w,text='Alice',font='Arial', width=10, height=4,activebackground='lightblue',activeforeground='blue',command=show_alice)
b1.pack(side='left',padx=100,pady=10)
b2.pack(side='left',padx=40,pady=20)
w.mainloop()
```

### 34 b - csv to Barplot

```python
import csv
data=[
    ['college','science','commerce','humanities'],
    ['A',22,40,53],
    ['B',54,23,60],
    ['C',44,90,20]
]
with open('university.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerows(data)

print('csv file saved successfully')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('university.csv')
colleges=df['college']
science=df['science']
commerce=df['commerce']
humanities=df['humanities']
bw=0.2
x=np.arange(len(colleges))
plt.figure(figsize=(10,6))
plt.bar(x,science,width=bw,label='science',color='blue')
plt.bar(x+bw,commerce,width=bw,label='commerce',color='red')
plt.bar(x+2*bw,humanities,width=bw,label='humanities',color='green')
plt.xlabel('Colleges')
plt.ylabel('Number of courses')
plt.xticks(x+bw,colleges)
plt.legend()
plt.grid(axis='y',linestyle='-',alpha=0.7)
plt.show()
```

### 34 a - Histogram

```python
import matplotlib.pyplot as plt

min_temps = [15, 16, 14, 17, 15, 18, 19, 13, 14, 15, 16, 18, 17, 15, 14, 
             19, 20, 21, 17, 18, 16, 14, 15, 17, 19, 21, 22, 23, 18, 20]
max_temps = [28, 30, 27, 29, 31, 32, 30, 29, 28, 27, 29, 31, 32, 33, 28, 
             27, 29, 30, 33, 34, 32, 30, 29, 28, 31, 33, 34, 35, 30, 32]
plt.figure(figsize=(10,5))
plt.hist(min_temps, bins=6, alpha=0.7, label='min temperature', color='yellow', edgecolor='green')
plt.hist(max_temps, bins=6, alpha=0.7, label='max temperature', color='red', edgecolor='green')
plt.xlabel('temperature(degree C)')
plt.ylabel('frequency(days)')
plt.legend()
plt.show()
```

### 33 b - Linear plot

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    'Name':['Shalini','Gaurav','Bhavya','Divansh'],
    'Age':[25,31,29,28],
    'Salary':[32000,35000,37000,27000]
}
emp_df=pd.DataFrame(data)
name=emp_df['Name']
age=emp_df['Age']
plt.plot(name,age,'r',linestyle='-',marker='d',markeredgecolor='b')
plt.title('Comparison of ages of employees')
plt.xlabel('Names of employees')
plt.ylabel('Ages of employees')
plt.show()
```
